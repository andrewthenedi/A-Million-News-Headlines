# Exploratory Data Analysis

```python
# https://stackoverflow.com/a/47091490/4084039
import re

def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

##### Combining all the headlines 

```python
from tqdm import tqdm_notebook
preprocessed_headlines = []
# tqdm is for printing the status bar
for sentence in tqdm_notebook(raw_data['headline_text'].values):
    sentence = re.sub(r"http\S+", "", sentence)
    sentence = BeautifulSoup(sentence, 'lxml').get_text()
    sentence = decontracted(sentence)
    sentence = re.sub("\S*\d\S*", "", sentence).strip()
    sentence = re.sub('[^A-Za-z]+', ' ', sentence)
    # https://gist.github.com/sebleier/554280
    sentence = ' '.join(e.lower() for e in sentence.split() if e.lower() not in stopwords)
    preprocessed_headlines.append(sentence.strip())
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import string

STOPWORDS = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean_text(text):
    stop_not = " ".join([i for i in text.lower().split() if i not in STOPWORDS])
    punctuation_free = "".join([i for i in stop_not if i not in exclude])
    normalized = " ".join(lemma.lemmatize(word) for word in punctuation_free.split())
    return normalized

document_cleaned = [clean_text(text).split() for text in preprocessed_headlines]
```

##### Define helper functions

```python
def get_top_n_words(n_top_words, count_vectorizer, text_data):
    '''
    returns a tuple of the top n words in a sample and their 
    accompanying counts, given a CountVectorizer object and text sample
    '''
    vectorized_headlines = count_vectorizer.fit_transform(text_data.values)
    vectorized_total = np.sum(vectorized_headlines, axis=0)
    word_indices = np.flip(np.argsort(vectorized_total)[0,:], 1)
    word_values = np.flip(np.sort(vectorized_total)[0,:],1)
    
    word_vectors = np.zeros((n_top_words, vectorized_headlines.shape[1]))
    for i in range(n_top_words):
        word_vectors[i,word_indices[0,i]] = 1

    words = [word[0].encode('ascii').decode('utf-8') for 
             word in count_vectorizer.inverse_transform(word_vectors)]

    return (words, word_values[0,:n_top_words].tolist()[0])
```

##### Plot the top 20 words

```python
reindexed_data = raw_data['cleaned_headline_text']
reindexed_data.index = raw_data['publish_date']

count_vectorizer = CountVectorizer(stop_words='english')
words, word_values = get_top_n_words(n_top_words=20,
                                     count_vectorizer=count_vectorizer, 
                                     text_data=reindexed_data)

fig, ax = plt.subplots(figsize=(16,8))
ax.bar(range(len(words)), word_values);
ax.set_xticks(range(len(words)));
ax.set_xticklabels(words, rotation='vertical');
ax.set_title('Top words in headlines dataset (excluding stop words)');
ax.set_xlabel('Word');
ax.set_ylabel('Number of occurences');
plt.show()
```

##### Part-of-Speech Data Distribution

```python
tagged_data = [TextBlob(reindexed_data[i]).pos_tags for i in range(reindexed_data.shape[0])]
tagged_data_df = pd.DataFrame({'tags':tagged_data})
word_counts = [] 
pos_counts = {}
for headline in tagged_data_df[u'tags']:
    word_counts.append(len(headline))
    for tag in headline:
        if tag[1] in pos_counts:
            pos_counts[tag[1]] += 1
        else:
            pos_counts[tag[1]] = 1
            
print('Total number of words: ', np.sum(word_counts))
print('Mean number of words per headline: ', np.mean(word_counts))
```

##### Word Length

```python
y = stats.norm.pdf(np.linspace(0,14,50), np.mean(word_counts), 
                   np.std(word_counts))

fig, ax = plt.subplots(figsize=(18,8))
ax.hist(word_counts, bins=range(1,14), density=True);
ax.plot(np.linspace(0,14,50), y, 'r--', linewidth=1);
ax.set_title('Headline Word Lengths');
ax.set_xticks(range(1,14));
ax.set_xlabel('Number of words');
plt.show()
```

##### Distribution of the Part-of-Speech Data

```python
pos_sorted_types = sorted(pos_counts, key=pos_counts.__getitem__, 
                          reverse=True)
pos_sorted_counts = sorted(pos_counts.values(), reverse=True)

fig, ax = plt.subplots(figsize=(18,8))
ax.bar(range(len(pos_counts)), pos_sorted_counts);
ax.set_xticks(range(len(pos_counts)));
ax.set_xticklabels(pos_sorted_types);
ax.set_title('Part-of-Speech Tagging for Headlines Corpus');
ax.set_xlabel('Type of Word');
```

# Topic Modelling Using Gensim Library

```python
bigram = gensim.models.Phrases(document_cleaned, min_count=5, 
                               threshold=100) 
bigram_mod = gensim.models.phrases.Phraser(bigram)
def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]
data_words_bigrams = make_bigrams(document_cleaned)
```

##### Creating a term dictionary of the corpus, which every unique term is being assigned to an index

```python
dictionary = corpora.Dictionary(data_words_bigrams)
```

##### Converting a list of headlines (corpus) into a headline term matrix using the  prepared dictionary 

```python
doc_term_matrix = [dictionary.doc2bow(text) for text in data_words_bigrams]
```

### LDA (Latent Dirichlet Allocation)

##### Build the LDA model

```python
from gensim.models import LdaModel
lda_model = LdaModel(corpus=doc_term_matrix,
                     id2word=dictionary,
                     num_topics=10)
```

```python
# Compute the coherence score for the LDA model using 'u_mass'
coherence_model_lda = CoherenceModel(model=lda_model, 
                                     texts=data_words_bigrams, 
                                     dictionary=dictionary, 
                                     coherence='u_mass')
coherence_lda_u_mass = coherence_model_lda.get_coherence()
print('\nCoherence Score using u_mass: ', coherence_lda_u_mass)
```

```python
from tqdm import tqdm
coherenceList_umass_lda = []
coherenceList_cv = []
num_topics_list = np.arange(10,21)
for num_topics in tqdm(num_topics_list):
    lda_model = LdaModel(corpus=doc_term_matrix,
                         id2word=dictionary,
                         num_topics=num_topics)
    
    cm = CoherenceModel(model=lda_model, 
                        corpus=doc_term_matrix, 
                        texts=data_words_bigrams, 
                        dictionary=dictionary, 
                        coherence='u_mass')
    coherenceList_umass_lda.append(cm.get_coherence())
```

```python
# Number of topics for the LDA model using coherence 'u_mass'

plotData = pd.DataFrame({'Number of topics':num_topics_list,
                         'CoherenceScore_umass':coherenceList_umass_lda})
f,ax = plt.subplots(figsize=(10,6))
sns.set_style("darkgrid")
sns.pointplot(x='Number of topics',y= 'CoherenceScore_umass',data=plotData)
plt.title('Topic coherence')
```

### Latent Semantic Indeixing (LSI)

```python
from gensim.models import LsiModel
lsi_model = LsiModel(corpus=doc_term_matrix,
                     id2word=dictionary,
                     num_topics=10)
```

```python
# Compute the coherence score for the LSI model using coherence 'u_mass'
coherence_model_lsi = CoherenceModel(model=lsi_model, 
                                     texts=data_words_bigrams, 
                                     dictionary=dictionary, 
                                     coherence='u_mass')
coherence_lsi_u_mass = coherence_model_lsi.get_coherence()
print('\nCoherence Score using u_mass: ', coherence_lsi_u_mass)
```

```python
from tqdm import tqdm
coherenceList_umass_lsi = []
num_topics_list = np.arange(10,21)
for num_topics in tqdm(num_topics_list):
    lsi_model = LsiModel(corpus=doc_term_matrix,
                         id2word=dictionary,
                         num_topics=num_topics)
    
    cm = CoherenceModel(model=lsi_model, 
                        corpus=doc_term_matrix, 
                        texts=data_words_bigrams, 
                        dictionary=dictionary, 
                        coherence='u_mass')
    coherenceList_umass_lsi.append(cm.get_coherence())
```

```python
# Number of topics for the LSI model using coherence 'u_mass'

plotData = pd.DataFrame({'Number of topics':num_topics_list,
                         'CoherenceScore_umass':coherenceList_umass_lsi})
f,ax = plt.subplots(figsize=(10,6))
sns.set_style("darkgrid")
sns.pointplot(x='Number of topics',y= 'CoherenceScore_umass',data=plotData)
plt.title('Topic coherence')
```

### HDP (Hierarchical Dirichlet Process)

```python
from gensim.models import HdpModel
hdp_model = HdpModel(corpus=doc_term_matrix,
                     id2word=dictionary)
```

```python
# Compute the coherence score for the HDP model using coherence 'u_mass'
coherence_model_hdp = CoherenceModel(model=hdp_model, 
                                     texts=data_words_bigrams, 
                                     dictionary=dictionary, 
                                     coherence='u_mass')
coherence_hdp_u_mass = coherence_model_hdp.get_coherence()
print('\nCoherence Score using u_mass: ', coherence_hdp_u_mass)
```

# Evaluating All Four Models Performance

```python
temp = pd.DataFrame({'Topic_Model':['LDA','LSI','NMF','HDP'],
                     'Coherence_Score':[
                         abs(max(coherenceList_umass_lda)),
                         abs(max(coherenceList_umass_lsi)),
                         abs(min(coherenceList_umass_nmf)),
                         abs(coherence_hdp_u_mass)]})
plt.bar(temp.Topic_Model,temp.Coherence_Score)
plt.show()
```

```python
# We can conclude that LSI peforms the best performance for this case
# , since it has the highest coherence score
# Hence, we pick LSI as the optimal algorithm
```
